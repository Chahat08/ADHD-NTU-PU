```python
file_path = r"C:\Users\CHAHAT KALSI\Desktop\adhd\datasets\eeg-alcoholic\small\extract\a_1_co2a0000364\co2a0000364.000.txt"
file = open(file_path, 'r')
#print(file.read())
```

## Information about the file

```python
i = 0
data = {}
for line in file.readlines():
    # only looking at channel 0 for now.
    if 'FP1' not in line and i>4: break
    if i==0:
        print('PATIENT IDENTIFIER: ', line.strip()[2:-3], end=', ')
        if line.strip()[2:][3]=='a': print('Alcoholic')
        else: print('Normal Control')
    elif i==1 or i==2 or i==3:
        print(line.strip()[2:])
    elif i==4:
        l = line.strip()[2:]
        print('ELECTRODE NAME: ', l.split(' ')[0], ', CHANNEL NUMBER:', l.split('FP1')[1])
    else:
        l = line.strip().split(' ')
        data[int(l[2])]=float(l[3])
    i+=1
file.close()
```

# PLOTTING ALL THE ELECTRODES

```python
file_path = r"C:\Users\CHAHAT KALSI\Desktop\adhd\datasets\eeg-alcoholic\small\extract\a_1_co2a0000364\co2a0000364.000.txt"
file = open(file_path, 'r')
```

```python
ls=[]
```

```python
i=0
n=''
vals=[]
for line in file.readlines():
    if i <= 3: i+=1 #just ignore the first 4 lines
    else:
        if '#' in line:
            if vals!=[]:
                ls.append(vals.copy())
                vals.clear()
            ls.append(line.split(' ')[1])
            n=line.split(' ')[1]
        else:
            l = line.strip().split(' ')
            vals.append(float(l[3]))
ls.append(vals)
file.close()
```

```python
ls[:2]
```

```python
title=''
idx=0
for x in ls:
    if '' in x: title=x
    else:
        plt.figure(idx)
        plt.plot(range(len(x)), x)
        plt.title(title)
        idx+=1
```

```python
plt.close('all')
```
